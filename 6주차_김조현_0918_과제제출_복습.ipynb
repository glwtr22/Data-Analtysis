## 데이터 시각화(EDA, Exploratory Data Analysis)

시각화 -> 새로운 인사이트 발견(데이터의 특징)
- 필수 과정이며, 데이터 분석 진행 중 초반에 시각화를 통해 새로운 인사이트를 찾는다.
- 시각화를 진행하기 위해서는 시각화의 목적이 분명해야 한다.
- 시각화의 목적이 정확해야 의미 부여가 가능하다.
- 시각화는 "간결"해야 한다.

#### 그래프 기초 / 환경 설정

```python
from matplotlib import pyplot as plt
```

```python
# 산점도 시각화
plt.scatter([1,2,3], [4,5,6])   # 첫 번째 배열 x축, 두 번째 배열 y축
```

```python
plt.scatter([1, 2, 3], [4, 5, 6])
plt.scatter([7, 8, 9],[1, 2, 6])  # 두 개를 그리면 다른 색으로 표현됨
```

#### 그래프의 제목 및 축 이름 설정

```python
plt.scatter([1,2,3],[4,5,6])
plt.title('Hi')  # 그래프의 제목, 그래프 상단 중앙에 표시됨
plt.xlabel('x hi')
plt.ylabel('hi')
```

#### 축 범위 설정
- xlim x축 설정
- ylim y축 설정
- left(해당 값 만큼 이동해서 시작) : x축에서의 설정 / top(최댓값) : y축에서의 설정

```python
plt.scatter ([1,2,3],[4,5,6])
plt.xlim(0, 2)
plt.ylim(3, 5)
```

```python
plt.scatter([1,2,3],[4,5,6])
plt.xlim(left = 0.5)
plt.ylim(top = 10)
```

```python
plt.scatter([1,2,3], [4,5,6])
```

```python
plt.scatter([1,2,3],[4,5,6])
plt.xlim(right = 1.5) 
plt.ylim(bottom = 100)  # xlim right 인자와 ylim bottom 인자가 잘 이해가 안됨
```

#### 범례 설정하기
- legend 함수 사용하기
- label로 범례의 이름 설정하기

```python
plt.scatter([1,2,3],[4,5,6], label = 'x')
plt.scatter([-1, -2, -3], [4, 5, 6], label = 'y')
plt.legend()
```

- 범례의 위치를 바꿔주는 인자
- best
- center
- right
- upper left
- upper center
- upper right
- lower left
- lower center
- lower right

```python
plt.scatter([1,2,3],[4,5,6], label = 'x')
plt.scatter([-1,-2,-3],[4,5,6], label = 'y')
# plt.legend(loc = 'upper right')
# plt.legend(loc = 'lower center')
plt.legend(loc = 'best')
```

## 그래프의 크기 및 스타일 설정하기
- figure 함수의 figsize

```python
plt.figure(figsize = (15, 5))
# 말그대로 출력했을 때 눈으로 보여지는 길이(비율)을 조절해줌
plt.scatter([1,2,3],[4,5,6], label = 'x')
```

## 다양한 그래프 그리기

#### 선 그래프 그리기

```python
plt.plot([1,2, 3],[4,5, 6], label = 'line x')   # scatter 대신에 plot 사용
plt.plot([1,2,3],[7,8,9], label = 'line y')
plt.legend(loc = 'upper left')
```

#### 그래프 색상 설정하기

- color : black, gray, yellow, brown, blue, lightblue
- 구글링 통해 더 많은 색상 확인

```python
plt.plot([1,2,3],[4,5,6],color ='lightgreen',label ='line x')
plt.plot([1,2,3],[7,8,9],color = 'lightblue', label ='line y')
plt.legend(loc='center')
```

#### 선 스타일, 두깨 설정
- linestyle (solid, dashed, dashdot, dotted)
- linewidth ()

```python
plt.plot([1,2,3],[4,5,6],linestyle='dashdot',linewidth=8,color ='brown',label ='line x')
plt.plot([1,2,3],[7,8,9],linestyle='dotted',linewidth=3,color = 'lightblue', label ='line y')
plt.legend(loc='center')
```

#### 마커 설정하기
- 마커의 종류는 다양하다 (구글링으로 확인 가능), marker 인자로 설정
- 마커 사이즈 markersize로 설정 가능

```python
plt.plot([1,2,3],[4,5,6],marker = '^',markersize=10,color ='brown',label ='line x')
plt.plot([1,2,3],[7,8,9],marker = 'x',markersize=10,color = 'lightblue', label ='line y')
plt.legend(loc='center')
```

```python
plt.scatter([1,2,3],[4,5,6],marker = 'H',s=40, label ='x')
plt.scatter([-1,-2,-3],[4,5,6],marker = 'd',s=80,label ='y')
# plot 함수와 인자가 거의 유사, markersize 인자는 s로 입력해야 함
```

#### 투명도 설정

데이터가 겹쳐져 있을 때 분포를 쉽게 알아보기 위해 사용

```python
import numpy as np
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y)
```

```python
plt.scatter(x, y, alpha = 0.1)  # alpha 값으로 투명도 조정
```
