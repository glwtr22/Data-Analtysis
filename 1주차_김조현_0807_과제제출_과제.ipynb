### 데이터 프레임

- 판다스는 파이썬을 활용해 데이터 분석을 하기 위해서 사용하는 필수적인 패키지이다.
- 판다스 데이터 프레임을(Pandas DataFrame)을 활용하면 데이터를 표(Table) 형태로 처리할 수 있어서 수집된 데이터의 전처리 등 각종 데이터 핸들리을 쉽고 편하게 할 수 있다.

```python
import pandas as pd
```

### 데이터에 대한 기초 이해
- head() : 데이터프레임에서 앞의 5행 출력
- tail() : 데이터프레임에서 뒤의 5행 출력
- shape : 데이터프레임의 행과 열의 개수를 튜플로 반환
- info() : 데이터의 변수 속성 관련 정보를 출력
- describe() : 데이터의 열 별 요약 통계량 출력

pandas의 데이터 타입에는 6가지가 있다.
- int64 : 정수형
- float64 : 실수형
- bool : 부울형
- datetime64 : 날짜 표현
- category : 카테고리형
-  object : 문자열 or 복합형

<font color=green>catetgory형 타입이란?
: 데이터 프레임의 열에서 특정한 형태의 데이터가 반복되는 경우 사용한다. 예를 들어 성별(남성, 여성), 나이(10대, 20대, ...)와 같이 특정 구간의 데이터가 반복되는 경우를 말한다.</font>

```python
df_1.describe()
```

요약 통계량은 값의 분포를 요약해주어 많은 양의 정보를 단순하게 만들어 준다.

- count : 빈도, 값의 개수
- mean : 평균
- std : 표준 편차
- min : 최소값
- 25% : 제1 사분위수(또는 하위 사분위수)
- 50% : 중앙값
- 75% : 제3 사분위수(또는 상위 사분위수)
- max : 최대값

수치형 데이터 <-> 범주형 데이터(category)

문자열 데이터는 object 자료형으로 나타낸다.

일반적인 문자열을 갖는 열은 object로 사용하고, 값의 종류가 제한적일 때 category를 사용하면 된다.

(예를 들어, 아침 식사 여부에 대한 열이라면 값을 0 = 먹지 않음, 1 = 먹음으로 두 종류를 갖게 되는데, 아침 식사 종류에 대한 열이라면 샐러드, 소고기, 바나나, 샌드위치, 마라탕 등 다양한 값을 가질 수 있게 된다.

아침 식사 여부는 category형으로, 아침 식사 종류는 object형으로 지정해주면 효율적으로 데이터 프레임을 관리할 수 있다.)

### 넘파이를 이용한 데이터 값의 변경

넘파이는 수치해석용 파이썬 패키지이다.

다차원 배열 자료구조 클래스인 ndarray 클래스를 지원하여,

벡터와 행렬을 사용하는 선형대수 계산에 주로 사용된다.

```python
import numpy as np
```

```python
df_1['test'] = np.where(df_1['cty']>=20, 1, 0)  # 첫 번째 인자로 들어온 조건식이 참일 경우 두 번째 인자의 요소를 반환, 거짓일 경우 세 번째 인자의 요소를 반환
```

```python
df_1['test'].value_counts()
```
