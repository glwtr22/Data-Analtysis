#### 시계열 데이터프레임 만들기
시계열 데이터란? : 일정한 시간 간격으로 배치된 데이터들의 수열을 말한다.

- datetime : 판다스에서 제공하는 시계열 데이터 형식
- 시계열 데이터는 Dtype이 꼭 datetime으로 변경되어야 한다.
- 이유는 datetime 문법이 따로 존재하고, 이 문법을 적용시키려면 datetime으로 설정되어있어야 한다.

```python
ts_df = pd.read_csv('resource/country_timeseries.csv')
```

```python
ts_df
```

```python
ts_df.info()
```

```python
ts_df['Date'] = pd.to_datetime(ts_df['Date']) # "Date" 칼럼을 datetime 형식으로 변경
```

```python
ts_df.info()
```

```python
ts_df[['Date']]
```

```python
ts_df
```

datetime 데이터는 년 월 일이 - (대쉬)로 구분되는 형태를 띈다.
