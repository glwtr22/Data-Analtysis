```python
import pandas as pd
```

### groupby, agg

```python
df_1 = df.groupby('manufacturer')['displ'].mean().to_frame()
```

```python
df_1
```

```python
df_1.reset_index()
```

### concat

```python
df1 = df[['manufacturer','cyl']]
df2 = df[['manufacturer','drv']]
```

```python
pd.concat([df1, df2], axis = 0)
```

### merge

```python
pd.merge(df1, df2, how='left', on=['manufacturer'])
```

```python
df1 = df[['drv','fl','category']]
df2 = df[['trans', 'fl', 'category']]
```

```python
pd.merge(df1,df2, how='inner', on=['category','fl'])
```
