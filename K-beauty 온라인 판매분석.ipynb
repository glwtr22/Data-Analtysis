### 분석과 시각화를 위한 tidy data

```python
df = df_raw.melt(id_vars=["국가(대륙)별", "상품군별", "판매유형별"], var_name = "기간", value_name = "백만원")
df
```

### 데이터 전처리 - 문자열 분리, 데이터 타입 변경

###### 1. 기간에서 연도/분기를 분리하기

```python
int("2019 4/4 p".split()[0])
```

```python
df['연도'] = df["기간"].map(lambda x : int(x.split('.')[0]))
df.head()
```

```python
df['분기'] = df['기간'].map(lambda x : int(x.split('.')[1].split('/')[0]))
df.head()
```

###### 2. 금액을 수치데이터로 표현하기 위한 데이터타입 변경

```python
df['백만원'] = df['백만원'].replace('-', pd.np.nan).astype(float)
df.head()
```

###### 3. 필요 없는 데이터 제거

```python
df.info()
```

```python
df = df[(df['국가(대륙)별'] != "합계") & (df['상품군별'] != '합계')].copy()
df.info()
```

```python
df.isnull().sum()
```

###### 3. 의류 패션 관련 상품 온라인 쇼핑 해외직접판매액

```python
df["상품군별"]
```

```python
df_fashion = df[(df["상품군별"] == "의류 및 패션 관련상품") & (df["판매유형별"] == "계")].copy()
df_fashion.head()
```

```python
df_fashion2 = df[(df["상품군별"] == "의류 및 패션 관련상품") & (df["판매유형별"] != "계")].copy()

plt.figure(figsize = (15, 4))
plt.xticks(rotation = 30)
sns.lineplot(data = df_fashion2, x = "기간", y = "백만원", hue = "판매유형별", ci = None)
```

###### 4. 데이터 집계하기

```python
# 피봇 테이블로 국가별 연도별 합계 금액을 표 형태로 구해보자 !
result = df_fashion.pivot_table(index = "국가(대륙)별", columns = "연도", values = "백만원", aggfunc = "sum") # default > mean
result
```

###### 5. 연산결과를 시각적으로 확인

```python
plt.figure(figsize = (10,6))
sns.heatmap(result, cmap = "Greens", annot = True, fmt = ".0f") # 숫자가 많은 피봇 테이블을 한눈에 확인 가능 !!
```
