#### 과제 mpg 데이터를 이용해서 배운 내용 적용하기!
- 모든 열/칼럼에 대해서 한 번씩 다 적용하기
- 분석가가 찾고자 하는 질문 -> 해당 질문 추출할 수 있는 코드 작성
- 인사이트 주석으로 정리
- ex) audi a4에 대한 연비 요약 통계
- 각 열/칼럼 당 최소 5개 이상 질문 만들기
- 랜덤 발표 진행

```python
import pandas as pd
```

```python
df = pd.read_csv('resource/mpg.csv')
```

```python
df
```

##### MPG : Miles Per Gallon(연비)

- manufacturer : 제조사

- model : 모델

- displ : 배기량

- year : 제조 연도

- cyl :실린더 개수

- trans : 차량 내부의 변속기 유형(수동/자동)

- drv : 구동 방식 (f = 전륜 구동 / r = 후륜 구동 / 4 = 사륜 구동)

- cty : 도시 연비

- hwy : 고속도로 연비

- fl : fuel (d : diesel, r : regular, p : premium, e : ethanol)

- category : 분류

#### 1 . manufacturer

```python
# 1. 제조사 별 자동차 개수
df['manufacturer'].value_counts().to_frame()
```

```python
# 2. volkswagen에서 제조한 자동차들에 대한 정보 출력하기
df.query('manufacturer == "volkswagen"')
```

```python
# 3. 제조사 별 도시 연비 평균
df.groupby('manufacturer').agg(cty_mean = ('cty','mean'))
```

```python
# 4. 제조사 별 실린더 최대 개수
df.groupby('manufacturer').agg(cyl_mean = ('cyl','max'))
```

```python
# 5. 제조사 별 차량 출시년도의 최솟값
df.groupby('manufacturer').agg(earliest_year = ('year','min'))
```

#### 2. model

```python
# 1. audi의 차 모델 종류 파악하기
df.query('manufacturer == "audi"')['model'].value_counts()
```

```python
# 2. audi의 a6 quattro 모델의 실린더 개수 파악
df.query('manufacturer == "audi"' and 'model == "a6 quattro"')['cyl'].to_frame()
```

```python
# 3. jeep에서 제조한 차들의 구동 방식 파악하기
df.query('manufacturer == "jeep"')['fl'].to_frame()
```

```python
# 4. jeep에서 제조한 차들의 구동 방식별 개수
df.query('manufacturer == "jeep"')['fl'].value_counts().to_frame()
```

```python
# 5. jeep에서 제조한 차들의 구동 방식별 개수 중 최댓값
df.query('manufacturer == "jeep"')['fl'].value_counts().max()
```

#### 3. displ

```python
# 1. 배기량이 가장 높은 차 출력
df['displ'].max()
```

```python
# 2. honda의 배기량 평균값
df.query('manufacturer == "honda"')['displ'].mean()
```

```python
# 3. 각 제조사 별 배기량 평균 구하기
df.groupby('manufacturer').agg(displ_mean = ('displ','mean'))
```

```python
# 4. 배기량이 5.0 이상인 자동차 개수
df_5 = df.query('displ >= 5.0')
len(df_5)
```

```python
# 5. 배기량과 실린더 수의 관계성 살펴보기
df[['displ','cyl']].sort_values('displ')
```

#### 4. year

```python
# 1. 19세기에 출시된 자동차 개수와 20세기에 출시된 자동차 개수
n1 = len(df.query('year<2000'))
n2 = len(df.query('year>=2000'))
print("19세기에 출시된 자동차 개수:", n1, ", 20세기에 출시된 자동차 개수:", n2)
```

```python
# 2. 19세기에 만들어진 차들의 실린더 개수 평균
df.query('year<2000')['cyl'].mean()
```

```python
# 3. 20세기에 만들어진 차들의 실린더 개수 평균
df.query('year>=2000')['cyl'].mean()
```

```python
# 4. 19세기에 만들어진 차들의 변속기 별 개수
df.query('year<2000').groupby('trans').agg(trans_ct = ('trans','count'))
```

```python
# 5. 20세기에 만들어진 차들의 변속기 별 개수
df.query('year>=2000').groupby('trans').agg(trans_ct = ('trans','count'))
```

#### 5. cyl

```python
# 1. 실린더 개수의 종류 별 개수
df['cyl'].value_counts().to_frame()
```

```python
# 2. 실린더 개수에 따른 고속도로 연비 평균
c4 = df.query('cyl == 4')['hwy'].mean()
c5 = df.query('cyl == 5')['hwy'].mean()
c6 = df.query('cyl == 6')['hwy'].mean()
c8 = df.query('cyl == 8')['hwy'].mean()
print("실린더 4개 고속도로 연비 :", c4)
print("실린더 5개 고속도로 연비 :", c5)
print("실린더 6개 고속도로 연비 :", c6)
print("실린더 8개 고속도로 연비 :", c8)
```

```python
# 3. 실린더 개수에 따른 도시 연비 평균
c4 = df.query('cyl == 4')['cty'].mean()
c5 = df.query('cyl == 5')['cty'].mean()
c6 = df.query('cyl == 6')['cty'].mean()
c8 = df.query('cyl == 8')['cty'].mean()
print("실린더 4개 도시 연비 :", c4)
print("실린더 5개 도시 연비 :", c5)
print("실린더 6개 도시 연비 :", c6)
print("실린더 8개 도시 연비 :", c8)
```

#### 6. trans
