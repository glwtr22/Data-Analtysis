#### 데이터 병합

merge, concat(join 이용)

```python
df1 = pd.DataFrame({'A':[1,2,3], 'B':[4,5,6]})
df2 = pd.DataFrame({'A':[1,2,3], 'C': [7,8,9]})
```

```python
display(df1)
display(df2)

# 그냥 데이터 프레임 변수 써서 표 보는 것과 display 함수 사용하는 것의 차이점
# : 그냥 데이터 프레임 쓰면 마지막 변수만 보여지지만,
# : display 함수를 사용하면 여러 표를 한 번에 볼 수 있다!
```

#### concat()

데이터프레임을 물리적으로 이어 붙여준다!

```python
pd.concat([df1, df2], axis = 0) # axis = 0 : Default 값 / 행 방향(위 아래 방향)으로 데이터프레임 붙여준다.
```

```python
pd.concat([df1, df2], axis=1) # axis = 1 : 열 방향(좌우 방향)으로 데이터프레임을 붙여준다.
```

#### merge()

두 데이터프레임을 각 데이터에 존재하는 고유값(key)을 기준으로 병합할 때 사용한다.

pd.merge(df_left, df_right, how='inner', on=None)이 default이다.

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, how='left', on = 'A')
```

#### 교집합이 2개 이상인 경우

```python
df1 = pd.DataFrame({'A':[1,2,3], 'B':[4,5,6], 'C':[13,14,15]})
df2 = pd.DataFrame({'A':[1,2,3], 'B':[4,5,7], 'D':[10,11,12]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, how='inner', on=['A','B'])
```

#### 결측치 찾아보기

결측치란, 누락된 값, 비어 있는 값

```python
import numpy as np
```

```python
df = pd.DataFrame({'A':[1,2,3,4,5,np.nan], 'B':[7,8,np.nan, 9,10,11]})
```

NaN 값은 0의 개념이 아니라 어떤 값을 더하거나 빼거나 곱하거나 나눠도 다 NaN 값이다!

```python
df['A']+1
```

```python
df.isna()
```

```python
df.isna().sum()
```

#### 결측치 제거

```python
df.dropna() # df에서 NaN이 있는 행을 제거해줌
```

```python
# 내가 원하는 컬럼(B 열)의 결측치 제거
df.dropna(subset = ['B'])
```

```python
exam.loc[[3,7,12],['science']] = np.nan # 원하는 열의 인덱스 지정해서 NaN 값 대입 가능
```
